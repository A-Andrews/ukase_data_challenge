```python
from sklearn.manifold import TSNE
import pandas as pd
from os import path
import plotly.express as px
```

```python
data_file = "processed_ohe_False_zoonosis_dataset_full_full.csv"
```

```python
df = pd.read_csv(path.join("..", "cleaned_data", data_file))
```

```python
df.drop_duplicates(inplace=True)
# df = df[df["subtype"].isin(["H7N9", "H5N1"])]
y = df["label"]
colours = df["subtype"]
X = df.drop(columns=["label", "subtype", "gid", "src"])
```

```python
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
fig = px.scatter(x=X_tsne[:, 0], y=X_tsne[:, 1], color=y, opacity=0.1)
fig.update_layout(
    title="t-SNE visualization of Custom Classification dataset",
    xaxis_title="First t-SNE",
    yaxis_title="Second t-SNE",
)
fig.show()
```
